# Let's plot total rides across all the cities in the dataframe first

# Next is ride by city - also over time



```python
df_reset = df.reset_index()

df_melted = df_reset.melt(id_vars=['date'], var_name='City', value_name='Rides')


fig = px.line(df_melted, x='date', y='Rides', color='City', facet_col='City', facet_col_wrap=3,
              title='Rides by City Over Time',
              height=900,  
              width=1200)  


fig.update_yaxes(title_text=None)


fig.update_layout(yaxis_title_text='Number of Rides')
fig.update_layout(
    margin=dict(l=50, r=50, t=50, b=50),
    title_x=0.5,
    title_font=dict(size=20)
)

fig.show()
```

# Weekly patterns might also be interesting

```python
df = df.reset_index()
df['date'] = pd.to_datetime(df['date'])

df['day_of_week'] = df['date'].dt.dayofweek

df_melted = df.melt(id_vars=['date', 'day_of_week'], 
                    var_name='City', 
                    value_name='Rides')
weekly_pattern = df_melted.groupby(['day_of_week', 'City'])['Rides'].mean().reset_index()


fig = px.bar(weekly_pattern, 
             x='day_of_week', 
             y='Rides', 
             color='City', 
             facet_col='City', 
             facet_col_wrap=3,
             labels={'day_of_week': 'Day of Week', 'Rides': 'Average Number of Rides'},
             title='Average Rides by Day of Week for Each City',
             category_orders={'day_of_week': list(range(7))})

fig.update_xaxes(ticktext=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], 
                 tickvals=list(range(7)))
fig.update_layout(showlegend=False, 
                  height=800, 
                  width=1200)

fig.show()
```

### Most cities show a clear weekly pattern with higher ride numbers on weekends

### Sunday typically sees a slight dip compared to Saturday in most cities

### The implication for TweetX campaign could be the timing: For most cities, focusing the campaign on Thursdays and Fridays could capitalize on the build-up to peak weekend demand

-  It could also be important for cities like Poznan and Warsaw to emphasize weekend availability and nightlife-related messaging

# Next we also must look at monthly trends

```python
df['month'] = df['date'].dt.to_period('M')
```

```python
numeric_columns = df.select_dtypes(include=[np.number]).columns

monthly_data = df.groupby('month')[numeric_columns].sum().reset_index()
monthly_data['month'] = monthly_data['month'].dt.to_timestamp()

melted_data = monthly_data.melt(id_vars=['month'], var_name='City', value_name='Rides')

fig = px.line(melted_data, x='month', y='Rides', color='City',
              title='Monthly Trends of Rides by City',
              labels={'month': 'Month', 'Rides': 'Number of Rides'},
              line_shape='linear', render_mode='svg')

fig.update_layout(legend_title_text='City',
                  xaxis_title='Month',
                  yaxis_title='Number of Rides',
                  height=600, width=1000)

fig.show()
```

### Some spikes occur in April/ May for Warsaw and in July for Krakow. Otherwise there seems to be some seasonality around August for Gdansk

### Poznan seems to be pretty stable

# Implication for the campaign could be that as it is to start mid Oct we could be expecting drops due to general seasonality
- we could also allocate higher budgets to high-potential growth cities like Poznan
- for seasonal cities like Gdansk, consider dynamic budget allocation with higher spend during peak seasons.
